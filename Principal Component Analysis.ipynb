#### Consider a randomly generated data :
* Plot the data
* Use PCA to reduce 2d data to 1d
* Show the shape of original and reduced data

```python
# importing required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
```

```python
# creating a random data

rng = np.random.RandomState(1)
X = np.dot(rng.rand(2,2),rng.randn(2,200)).T
X
```

```python
# Shape of original data

X.shape
```

```python
# plotting the data using matplotlib 

plt.scatter(X[:,0],X[:,1])
plt.title("Distribution of the Data");
```

```python
# Reducing the dimension using PCA

Model = PCA(n_components = 1)
Model.fit(X)
New_X = Model.transform(X)
New_X
```

```python
# Shape of reduced data

New_X.shape
```

##### Conclusion:
    
   * Using the Principal Component Analysis Method we did the feature selection.
   * We reduced our 2- dimensional data (shape = (200,2)) to a 1-dimensional data (shape(200,1)).
